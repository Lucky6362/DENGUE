```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

```python
target = new_client.target
target_search = target.search('Dengue')
x = pd.DataFrame(target_search)
x
```

```python
x.shape
```

```python
selected=x.target_chembl_id[5]
selected
```

```python
activity=new_client.activity
res=activity.filter(target_chembl_id=selected).filter(standard_type='IC50')
```

```python
df1=pd.DataFrame(res)
df1
```

```python
df1.head()
```

```python
df1.standard_type.unique()
```

```python
df1.to_csv('bio-1.csv',index=False)
```

```python
df1.isnull().sum()
```

```python
df2=df1.dropna(subset=['standard_value'])
```

```python
df2.shape
```

```python
bio_class=[]
for i in df2.standard_value:
    if float(i)>=10000:
        bio_class.append('inactive')
    elif float(i)<=1000:
        bio_class.append('active')
    else:
        bio_class.append('intermediate')
```

```python
bio_class
```

```python
df3=df2[['molecule_chembl_id','standard_value','canonical_smiles']]
```

```python
df3.head()
```

```python
df3['bioactivity']=bio_class
df3.head()
```
